```python
df=df.sort_values(by=["IDENTIFICACION","PERIODO_COMERCIAL","fecha_liquidacion"])
```

```python
#las unicas variables que se repiten son monto colocado y fecha de liquidacion de estos
columnas_no_cambian=df.drop(columns=["MONTO_COLOCADO","fecha_liquidacion"]).columns
columnas_no_cambian
```

```python
df_colocacion_agrupado = (
    df.groupby(['IDENTIFICACION','PERIODO_COMERCIAL'], as_index=False)
      .agg({
          # Para las columnas categóricas, tomar el primer valor
          **{col: 'first' for col in columnas_no_cambian},
          # Para las columnas numéricas, aplicar la suma
          'fecha_liquidacion': 'last',
          'MONTO_COLOCADO': 'sum'
      })
)
```

```python
df_colocacion_agrupado
```

```python
df_colocacion_agrupado['PERIODO_COMERCIAL'] = pd.to_datetime(df_colocacion_agrupado['PERIODO_COMERCIAL'], format='%Y%m')
```

```python
df_colocacion_agrupado.PERIODO_COMERCIAL
```

```python
df_colocacion_agrupado['MES_PERIODO_COMERCIAL'] = df_colocacion_agrupado['PERIODO_COMERCIAL'].dt.to_period('M')
```

```python
#Agrupar por MES y contar la cantidad de registros
agrupado_mes = df_colocacion_agrupado.groupby('MES_PERIODO_COMERCIAL').size().reset_index(name='cantidad_registros')
```

```python
agrupado_mes
```

```python
df_colocacion_agrupado['APROBACION_MES_ANTERIOR'] = df_colocacion_agrupado['APROBACION_MES_ANTERIOR'].fillna('Ninguno')
df_colocacion_agrupado['PRODUCTO_MES_ANTERIOR'] = df_colocacion_agrupado['PRODUCTO_MES_ANTERIOR'].fillna('Ninguno')
```

```python
df_colocacion_agrupado['fecha_liquidacion']= df_colocacion_agrupado['fecha_liquidacion'].fillna('Ninguno')
```

```python
df_colocacion_agrupado.drop(columns=['CANAL_VENTA'], inplace=True)
```
